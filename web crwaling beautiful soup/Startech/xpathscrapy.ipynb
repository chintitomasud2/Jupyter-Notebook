```python
from lxml import html
import requests
import pandas as pd
price=[]
page = requests.get('https://www.startech.com.bd/networking/router?limit=100')
tree = html.fromstring(page.content)

links = tree.xpath("//div[@class='caption']/h4/a/@href")

title= tree.xpath("//div[@class='caption']/h4/a/text()")
price_1=tree.xpath("//div[@class='caption']//p[@class='price']/text()")
#print((price_1)[0])

for haire in price_1:
    price.append(haire.strip())
    
print(price)
    #print(pc)
#print("links",links)
#print("title",title)

data={'title':title,'links':links,'price':price}
```

```python
haire=pd.DataFrame(data)
haire=haire[['title','links','price']]
```
