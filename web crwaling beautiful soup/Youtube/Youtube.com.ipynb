```python
from bs4 import BeautifulSoup
import requests

url="https://www.youtube.com"
data=requests.get(url)
soup=BeautifulSoup(data.text,"html.parser")
```

```python
alllinks=soup.find_all('a')
```

```python
for links in alllinks:
    link=links.get('href')
    if link[0:3]=='/wa':
        print("https://www.youtube.com/"+link)
   
```
